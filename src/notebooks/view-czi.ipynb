```python
# imports
import czifile
import cv2
import matplotlib.pyplot as plt
import time
```

```python
# paths
DATA_PATH = '../../../../Data_DrosophiliaAxons/'
```

```python
with czifile.CziFile(DATA_PATH + '5dLL/GMRwhiteRNAi-w;GMRGal4-+;UASTubgfp-+_horizontal_5dLL_0.5AU_1A.czi') as czi:
    image = czi.asarray()
    metadata = czi.metadata()

print('Dataset Shape: ' + str(image.shape))
# ?, ?, channels, z-index, w, h, ?)

slice = image[0,0,0,20, :, :,0]
slice_normalized = cv2.normalize(slice, None, 0, 255, cv2.NORM_MINMAX).astype('uint8')
slice_rotated = cv2.rotate(slice_normalized, cv2.ROTATE_90_COUNTERCLOCKWISE)

plt.title("Slice")
plt.axis('off')
plt.imshow(slice_rotated, cmap='grey')
plt.show()
```
